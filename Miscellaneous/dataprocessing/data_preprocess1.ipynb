```python
import pandas as pd
data = pd.read_csv('C:/Users/Jixua/Desktop/plantDataset/utf8/combined.csv', encoding='latin-1')
```

```python
data['processed'] = data['price'].str.strip().str.split('$').str[1].str.split().str[0]
data = data.drop(columns=['total','3 payment','price','name1'])
data['num_columns'] = data['content'].apply(lambda x: len(x.split('\n\n')))
data
```

```python
import pandas as pd

# 假设你有一个名为 data 的 DataFrame，其中包含一个名为 "num_columns" 的列

# 初始化计数器
count = 0
df = pd.DataFrame()

# 使用 for 循环遍历 DataFrame 的行
for index, row in data.iterrows():
    # 检查 'num_columns' 列的值是否为 18 或 16
    if row['num_columns'] == 18 or row['num_columns'] == 16:
        count += 1
    else:
        # 删除不满足条件的行
        data = data.drop(index)

        # 如果值为 16，添加第三列和第四列为空列
    if row['num_columns'] == 16:
            content_value = data.loc[index, 'content']

            # 将指定行的 'content' 列的值按照 '\n\n' 进行拆分，并展开为多列
            split_result = content_value.split('\n\n')
            df_split_result = pd.DataFrame([split_result], columns=['name1', 'DISCLAIMER','name2','GENERAL INFORMATION','name3','SOIL REQUIREMENT', 
                                                                    'shopmore','name4','FEEDING FERTILIZER?REQUIREMENT',
                                                                   'shop','name5','LIGHT REQUIREMENT','name6','WATER REQUIREMENT','1','2'])
#             df_split_result.insert(df_split_result.columns.get_loc('name2'), 'new_column1', '')
#             df_split_result.insert(df_split_result.columns.get_loc('name2'), 'new_column2', '')
            # 将原始行和新的 DataFrame 进行列连接
            df = pd.concat([df, df_split_result], axis=0)
    if row['num_columns'] == 18:
            content_value = data.loc[index, 'content']

            # 将指定行的 'content' 列的值按照 '\n\n' 进行拆分，并展开为多列
            split_result = content_value.split('\n\n')
            df_split_result = pd.DataFrame([split_result], columns=['name1', 'DISCLAIMER','product','PRODUCT MEASUREMENT','name2','GENERAL INFORMATION','name3','SOIL REQUIREMENT', 
                                                                    'shopmore','name4','FEEDING FERTILIZER?REQUIREMENT',
                                                                   'shop','name5','LIGHT REQUIREMENT','name6','WATER REQUIREMENT','1','2'])
            # 将原始行和新的 DataFrame 进行列连接
            df = pd.concat([df, df_split_result], axis=0)
print(count)
print(len(data))
data.reset_index(drop=True, inplace=True)
df.reset_index(drop=True, inplace=True)

# 然后进行重新索引操作
data = pd.concat([data, df], axis=1)
data
```
